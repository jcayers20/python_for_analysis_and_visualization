```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
dfl = DataFrame({'key': ['X', 'Y', 'Z', 'X', 'Y'], 'data': range(5)})
dfl
```

```python
dfr = DataFrame({'group_data': [10, 20]}, index = ['X', 'Y'])
dfr
```

```python
pd.merge(dfl, dfr, left_on = 'key', right_index = True)
```

```python
df_left_hr = DataFrame({'key1': ['SF', 'SF', 'SF', 'LA', 'LA'], 'key2': [10, 20, 20, 30, 40], 'data_set': np.arange(5.)})
df_left_hr
```

```python
df_right_hr = DataFrame(np.arange(10).reshape((5, 2)),
                        index = [['LA', 'LA', 'SF', 'SF', 'SF'],[20, 10, 10, 10, 20]],
                        columns = ['c1', 'c2'])
df_right_hr
```

```python
pd.merge(df_left_hr, df_right_hr, left_on = ['key1', 'key2'], right_index = True)
```
