```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df1 = DataFrame({'key': ['X', 'Z', 'Y', 'Z', 'X', 'X'], 'data_set_1': np.arange(6)})
df1
```

```python
df2 = DataFrame({'key': ['Q', 'Y', 'Z'], 'data_set_2': [1, 2, 3]})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on = 'key')
```

```python
pd.merge(df1, df2, on = 'key', how = 'left')
```

```python
pd.merge(df1, df2, on = 'key', how = 'right')
```

```python
help(pd.merge)
```

```python
pd.merge(df1, df2, on = 'key', how = 'outer')
```

```python
df3 = DataFrame({'key': ['X', 'X', 'X', 'Y', 'Z', 'Z'], 'data_set_3': range(6)})
df3
```

```python
df4 = DataFrame({'key': ['Y', 'Y', 'X', 'X', 'Z'], 'data_set_4': range(5)})
df4
```

```python
pd.merge(df3, df4)
```

```python
df_left = DataFrame({'key1': ['SF', 'SF', 'LA'], 'key2': ['one', 'two', 'one'], 'left_data': [10, 20, 30]})
df_left
```

```python
df_right = DataFrame({'key1': ['SF', 'SF', 'LA', 'LA'], 'key2': ['one', 'one', 'one', 'two'], 'right_data': [40, 50, 50, 70]})
df_right
```

```python
pd.merge(df_left, df_right, on = ['key1', 'key2'], how = 'outer')
```

```python
pd.merge(df_left, df_right, on = 'key1')
```

```python
pd.merge(df_left, df_right, on = 'key1', suffixes = ('_left', '_right'))
```
